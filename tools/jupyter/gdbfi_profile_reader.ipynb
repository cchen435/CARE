```python
from pathlib2 import Path
import pandas as pd
```

```python
profile = 'gdbfi.profile'
data = pd.read_csv(profile, sep = ';')
```

```python
data
```

```python
dynamic_insts = data['executions'].sum()
memacc_insts = data.loc[((data['MemRead'] == 1) | (data['MemWrite'] == 1))]['executions'].sum()
print('Total Dynamic instructions: ', dynamic_insts)
print('Memory Access Dynamic Instructions: ', memacc_insts)
print('Ratio: ', memacc_insts * 1.0 / dynamic_insts)
```
